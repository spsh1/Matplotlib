# Line Plot Basics

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
print(x)
```

```python
y=2*x
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

# Adding Titles and Labels

```python
plt.plot(x,y)
plt.title("Line Plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# Changing Line Aethetics

```python
plt.plot(x,y,color="orange",linestyle=":",linewidth=3)
plt.title("Line Plot of a and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# Adding two lines in the same plot

```python
y1=2*x
y2=3*x
plt.plot(x,y1,color="g",linewidth=3)
plt.plot(x,y1,color="blue",linestyle=":",linewidth=5)
plt.grid(False)
```

# Adding subplots horizontally

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.show()
```

# Adding subplot Vertically

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.subplot(2,1,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.show()
```

# Pie chart

```python
Fruits = ["apple","mangoes","banana","litchi","orange"]
quantity = [53,43,12,97,80]

plt.pie(quantity, labels=Fruits, autopct="%0.1f%%",colors=["green","yellow","red","maroon","pink"])
plt.show()
```

# Donut Chart

```python
Fruits = ["apple","mangoes","banana","litchi","orange"]
quantity = [53,43,12,97,80]

plt.pie(quantity, labels=Fruits, autopct="%0.1f%%",colors=["green","yellow","red","maroon","pink"],radius=2)
plt.pie([5],colors="w",radius=1)
plt.show()
```

```python
r = np.linspace(0,10,100)            #Polar plot
theta = 2 * np.pi * r

plt.figure(figsize=(6,6))
plt.subplot(111, projection='polar')
plt.plot(theta,r)
plt.title('Polar Plot')
plt.show()
```

# Contor plot or space plot

```python
#generate some data
x=np.linspace(-3.0,3.0,100)
y=np.linspace(-3.0,3.0,100)
X,Y = np.meshgrid(x,y)
Z= np.sin(X**2 + Y**2)

#Create a contour plot
plt.contour(X, Y, Z)
plt.title('Contor Plot')  #title of the plot
plt.xlabel('X-axis')      #label for the x-axis
plt.ylabel('Y-axis')      #label for the y-axis
plt.colorbar(label='Z')   #add a color bar with label
plt.show()
```
